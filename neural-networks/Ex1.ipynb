```python
import numpy as np

# Number of training iterations
NUM_ITERATIONS = 50
# Threshold for 0/1 classification
THRESHOLD = 0.0
# Learning rate
LEARNING_RATE = 1e6

# Data (both training and testing)
input_data = np.array([
    [0, 0, 1],
    [1, 0, 1],
    [0, 1, 1],
    [1, 1, 1]
])
output_data = np.array([
    -1,
    -1,
    -1,
    1
])

# Create perceptron weights (random weights)
weights = np.random.randn(3)

# Train perceptron
for iteration in range(NUM_ITERATIONS):
    # Calculate predictions with current weights
    predictions = (np.dot(input_data, weights) > THRESHOLD).astype(np.int)
    # Turn into {-1,1}
    predictions = (predictions * 2) - 1

    # Calculate accuracy (not needed for training, but to track the learning progress)
    accuracy = np.mean(predictions == output_data)
    # Print the accuracy
    print("Iteration %d: Acc %f \t %s" % (iteration, accuracy, str(predictions)))

    # Update weights according to update rule
    weights = weights + LEARNING_RATE * np.mean((output_data - predictions)[:, None] * input_data, axis=0)

# Print weights for inspection
print(weights)
```
